<h2>Normalizacion de los datos</h2>

.<h3>Graficamos para ver la distribución de los datos</h3>


```python

# Contar el número de ocurrencias de cada valor en la columna 'sexo'
conteo = df['sexo'].value_counts()

# Calcular el porcentaje para cada valor de 'sexo'
porcentaje = (conteo / conteo.sum()) * 100

# Crear el gráfico de barras
fig, ax = plt.subplots()

# Graficar el total y porcentaje de cada valor en 'sexo'
conteo.plot(kind='bar', color='skyblue', ax=ax, label='Total')
porcentaje.plot(kind='bar', color='blue', ax=ax, secondary_y=True, label='Porcentaje (%)')

# Configurar etiquetas y leyendas
plt.xticks(rotation=0)
plt.ylabel('Total')
plt.xlabel('Sexo')
plt.legend(loc='upper right')
plt.title('Total y Porcentaje por Sexo')

# Mostrar el gráfico
plt.show()
```

```python

# Contar el número de ocurrencias de cada valor en la columna 'sexo'
conteo = df['rol'].value_counts()

# Calcular el porcentaje para cada valor de 'sexo'
porcentaje = (conteo / conteo.sum()) * 100

# Crear el gráfico de barras
fig, ax = plt.subplots()

# Graficar el total y porcentaje de cada valor en 'sexo'
conteo.plot(kind='bar', color='skyblue', ax=ax, label='Total')
porcentaje.plot(kind='bar', color='lightgreen', ax=ax, secondary_y=True, label='Porcentaje (%)')

# Mostrar los valores numéricos de porcentaje en cada barra
for i, v in enumerate(porcentaje):
    ax.text(i, v + 0.1, f'{v:.2f}%', color='black', ha='center')

# Configurar etiquetas y leyendas
plt.xticks(rotation=0)
plt.ylabel('Total')
plt.xlabel('Rol')
plt.legend(loc='upper right')
plt.title('Total y Porcentaje por Rol')

# Mostrar el gráfico
plt.show()
```
